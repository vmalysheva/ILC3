```r
### LOLA run on merged peakmatrix
### this script checks if PIRs (with various filtering) are enriched for TF binding sites
### Fig. 5B


library(LOLA)
library(data.table)
library(GenomicRanges)
library(ggplot2)
library(ggrepel)
library(stringr)

## loading LOLA regionDB

regionDB = loadRegionDB("~/LOLA_core_caches/LOLACore/hg38/")
```

```r
### loading the user set

chic_abc_merged <- fread("~/miniPCHiC/hILCs/rcogs/git_hILCs_COGS/ILC3_chicago_fres_5kb_abc_fres_extended_peakm.txt")

head(chic_abc_merged)
```

```r
## active/open PIRs user set

ilc_actopen_pirs <-  unique(chic_abc_merged[atac == "TRUE" | k27ac == "TRUE"][,c("oeChr", "oeStart", "oeEnd")])
ilc_actopen_pirs[,oeChr := paste0("chr", oeChr)]
ilc_actopen_pirs <- GRanges(ilc_actopen_pirs)

## all PIRs universe

all_pirs <-  unique(chic_abc_merged[,c("oeChr", "oeStart", "oeEnd")])
all_pirs[,oeChr := paste0("chr", oeChr)]
all_pirs <- GRanges(all_pirs)
```

```r
locResults = runLOLA(ilc_actopen_pirs, all_pirs, regionDB, cores=1)
ILC_actop_pirs <- locResults[qValue < 0.05, c("qValue", "pValueLog", "antibody", "oddsRatio", "support", "maxRnk", "filename")][order(qValue, decreasing=FALSE),]
fwrite(x = locResults, file = "~/miniPCHiC/hILCs/hILCs_run2/git_ILCs_LOLA/lola_ilc_fres_5kb_chic_fres_abc_actopen_pirs_vs_all_pirs.txt", sep = "\t", scipen = 999)
```

```r
ILC_actop_pirs <- ILC_actop_pirs[pValueLog > 2][!is.na(antibody)]
ILC_actop_pirs[,c("Gene_name","PY") := data.table(str_split_fixed(antibody,"_",2))]
```

```r
p_actop <- ggplot(ILC_actop_pirs, aes(x = oddsRatio, y = support, color = log(qValue,10))) +
                geom_point(size = 0.5) + 
                scale_color_gradient2(high = "grey",
                      mid = "blue",
                      low = "red",
                      space = "Lab",
                      guide = "colourbar",
                      aesthetics = "colour") +
                geom_text_repel(box.padding = 0.5, segment.alpha = 0.8, segment.size = 0.5, 
                                segment.color = "grey20", min.segment.length = 0, data = ILC_actop_pirs[order(qValue, decreasing = FALSE)],colour = "grey20", aes(label=Gene_name)) +
                theme(axis.text.x = element_text(size = 6),
                    axis.text.y = element_text(size = 6),
                    axis.title.y = element_text(size = 6, colour = "grey30"),
                    legend.title = element_text(size=6, colour = "grey30"),
                    legend.text = element_text(size=6, colour = "grey30"),
                    legend.key.width = unit(0.3, 'cm'),
                    legend.key.height = unit(0.3, 'cm'),
                    panel.background = element_rect(fill = "white"),
                    axis.line.x=element_line(color="grey30"),
                    axis.line.y=element_line(color="grey30")
                    )
p_actop
```

```r
### LOLA for PIRs of genes that were prioritised by multiCOGS
## select interactions of genes that were prioritised by multiCOGS

## load multiCOGS Crohn's dosease deLange hg38 (combinedInteractions_ALL)

delange_susie_res <- fread("/rds/general/project/lms-spivakov-analysis/live/HRJ_monocytes/hILCs/COGS_results/COGS_out/CD_deLange_ILCs_hg38_SuSIE_combinedInteractions_ALL//Annotated_COGS_scores_data.table.txt")

## active/open PIRs user set

ilc_actopen_susie_gene_pirs <-  unique(chic_abc_merged[atac == "TRUE" | k27ac == "TRUE"][baitName %in% delange_susie_res[cogs>=0.5]$gene][,c("oeChr", "oeStart", "oeEnd")])
ilc_actopen_susie_gene_pirs[,oeChr := paste0("chr", oeChr)]
ilc_actopen_susie_gene_pirs <- GRanges(ilc_actopen_susie_gene_pirs)

## all PIRs universe

all_actopen_susie_gene_pirs <-  unique(chic_abc_merged[atac == "TRUE" | k27ac == "TRUE"][baitName %in% delange_susie_res[cogs>=0]$gene][,c("oeChr", "oeStart", "oeEnd")])
all_actopen_susie_gene_pirs[,oeChr := paste0("chr", oeChr)]
all_actopen_susie_gene_pirs <- GRanges(all_actopen_susie_gene_pirs)

## run LOLA
locResults_susie = runLOLA(ilc_actopen_susie_gene_pirs, all_actopen_susie_gene_pirs, regionDB, cores=1)
ILC_actop_susie_pirs <- locResults_susie[qValue < 0.05, c("qValue", "pValueLog", "antibody", "oddsRatio", "support", "maxRnk", "filename")][order(qValue, decreasing=FALSE),]
fwrite(x = locResults_susie, file = "lola_actopen_susie_pirs_vs_allactopenppi_pirs.txt", sep = "\t", scipen = 999)

ILC_actop_susie_pirs <- ILC_actop_susie_pirs[pValueLog > 2][!is.na(antibody)]
ILC_actop_susie_pirs[,c("Gene_name","PY") := data.table(str_split_fixed(antibody,"_",2))]
fwrite(x = ILC_actop_susie_pirs, file = "lola_actopen_susie_pirs_vs_allactopenppi_pirs_signig.txt", sep = "\t", scipen = 999)
```
